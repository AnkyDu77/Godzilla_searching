# Matching up Prices and Txns Outputs Datasets

```python
import pandas as pd
from StrFloatConvert import strtofloatconvert as str_flt

print('All good! Go on!')
```

```python
raw_ds1 = pd.read_csv('.../Data/outs_number_dated(all_mit_ds).csv')
raw_ds2 = pd.read_csv('.../Data/additional_outs_dated(2018-02-09_2018-04-28).csv')
marked_sums_1 = raw_ds1.copy()
marked_sums_2 = raw_ds2.copy()

marked_sums_1.drop('blockID', axis=1,inplace=True)
marked_sums_1['Date']=pd.to_datetime(marked_sums_1['Date'])
marked_sums_1['out_sums'] = marked_sums_1['sum_outs']
marked_sums_1.drop('sum_outs', axis=1,inplace=True)
marked_sums_1
```

```python
full_marked_sums = pd.concat((marked_sums_1,marked_sums_2),axis=0).reset_index(drop=True)
full_marked_sums
```

```python
full_marked_sums['Date'] = pd.to_datetime(full_marked_sums['Date'])
prepared_blckch_data = full_marked_sums.groupby(pd.Grouper(key='Date', freq='D')).sum().reset_index()
prepared_blckch_data
```

```python
price_data=pd.read_csv('/Users/aaraybin/Documents/py4e/ML/Bitcoin/mysite/btc/data/btc_price_10_04-18.csv')
price_data
```

```python
price_data['Date'] = pd.to_datetime(price_data['Date'])
c=0
dates_differencies = []
for i in range(0,1091):
    difference = price_data['Date'][c+1] - price_data['Date'][c]
    if  difference.days != -1:
        dates_differencies.append(price_data['Date'][c])
        dates_differencies.append(price_data['Date'][c+1])
    c+=1
len(dates_differencies)
```

```python
prepared_blckch_data = prepared_blckch_data.sort_values(by='Date', ascending=False).reset_index(drop=True)
prepared_blckch_data = prepared_blckch_data.iloc[:len(price_data)]
prepared_blckch_data
```

```python
price_data = str_flt(price_data)
price_data['Date'] = prepared_blckch_data['Date']
price_data['out_sums'] = prepared_blckch_data['out_sums']
price_data
```
